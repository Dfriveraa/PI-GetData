```python
import pandas as pd
import matplotlib.pyplot as plt
import random
df1=pd.read_csv('Muestras5s/Encender.csv')
df2=pd.read_csv('Muestras5s/Encender.csv')
```

```python
cambios=df1[df1['Tiempo']=='Order'].index.tolist()
print(cambios)
```

```python
rnd1,rnd2=random.choices(cambios,k=2)
```

# Dividimos

```python
gesto1=df1[(df1.index>=rnd1-300)&(df1.index<rnd1)].astype(int)
gesto2=df2[(df2.index>=rnd2-300)&(df2.index<rnd2)].astype(int)
```

```python
gesto1[['Acx','Acy','Acz']].plot(title='Gesto 1')
gesto2[['Acx','Acy','Acz']].plot(title='Gesto 2')
```

```python
gesto1[['Gx','Gy','Gz']].plot()
gesto2[['Gx','Gy','Gz']].plot()
```
