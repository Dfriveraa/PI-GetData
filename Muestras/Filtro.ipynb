```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ruptures as rpt
import random
df1=pd.read_csv('Muestras-Daniel/Activacion.csv')
```

```python
gesto1=df1[(df1.index>=188)&(df1.index<371)].astype(int)
```

```python
gesto1[['Acx']].plot()
gesto1[['Acy']].plot()
gesto1[['Acz']].plot()
#
```

```python
# a=df1[(df1.index<30)&(df1.index>=20)].astype(int)[['Acx']]
frecuancias=0
x=np.array(gesto1['Acx'])
x_1=np.array(gesto1['Acx'])
plt.plot(x)
plt.show()
x=np.fft.rfft(x)
f=17
ones=np.ones(f)
zeros=np.zeros(np.size(x)-f)
c=np.concatenate((ones,zeros))
clean=x*c
x=np.fft.irfft(clean)
plt.plot(x)
```

```python

y=np.array(gesto1['Acy'])
z=np.array(gesto1['Acz'])

model = "l2"
algo = rpt.Window(width=20, model=model).fit(x)
my_bkps = algo.predict(n_bkps=5)
rpt.show.display(x, my_bkps, figsize=(10, 6))
plt.title('X filtrado')
plt.show()

algo = rpt.Window(width=20, model=model).fit(x_1)
my_bkps = algo.predict(n_bkps=5)
rpt.show.display(x_1, my_bkps, figsize=(10, 6))
plt.title('X original')
plt.show()
# model = "l2"
# algo = rpt.Window(width=20, model=model).fit(y)
# my_bkps = algo.predict(n_bkps=5)
# rpt.show.display(y, my_bkps, figsize=(10, 6))
# plt.title('Change Point Detection: Window-Based Search Method')
# plt.show()

# model = "l2"
# algo = rpt.Window(width=20, model=model).fit(z)
# my_bkps = algo.predict(n_bkps=5)
# rpt.show.display(z, my_bkps, figsize=(10, 6))
# plt.title('Change Point Detection: Window-Based Search Method')
# plt.show()
```
